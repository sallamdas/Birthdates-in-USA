```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads=pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
recent_grads.head()
recent_grads.tail()
recent_grads.describe()
recent_grads=recent_grads.dropna()
```

```python
raw_data_count=recent_grads.count()
print(raw_data_count)
```

```python
recent_grads.plot(x='Sample_size',y='Employed',kind='scatter',title='Employed vs. Sample_size',figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size',y='Median',kind='scatter',title='Employed vs. Sample_size',figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',title='Employed vs. Sample_size',figsize=(5,10))
```

```python
recent_grads.plot(x='Full_time',y='Median',kind='scatter',title='Employed vs. Sample_size',figsize=(5,10))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```
